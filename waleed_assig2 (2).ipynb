```python
Applied Data Science Assignment 2
Introduction
This dataset presents historical stock price information for Zomato,
a leading online food delivery and restaurant aggregator company.
The dataset is compiled with data collected over a specific time period,
showcasing the fluctuation in Zomato's stock prices over days, weeks,
or months, depending on the granularity of the dataset.
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("zomato.csv")

# Display the first few rows of the dataset
print(data)
print(data.head())
```

```python
# Histogram of Closing Prices

import matplotlib.pyplot as plt

# Plot a histogram of closing prices
plt.hist(data['Close'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()

```

```python
# Line Graph of Closing Prices over Time

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Plot a line graph of closing prices over time
plt.plot(data['Date'], data['Close'], color='green')
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot a heatmap of the correlation matrix
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Coreelation Between Variables

# Summary statistics for the dataset
summary_stats = data.describe()
print(summary_stats)

# Correlation between variables
correlation = data.corr()
print(correlation)
```
